```python
import pandas as pd
import numpy as np
```

```python
path ="Datasets/iris.data"
names=['sepal-lenght','sepal-width','petal-lenght','petal-width','Target']
irisdata =pd.read_csv(path, names=names)
irisdata.head(10)
```

```python
print("Iris data set dimensions : {}".format(irisdata.shape))
```

```python
irisdata.info()
```

```python
irisdata.isnull().sum()
irisdata.isna().sum()
```

```python
x=irisdata.iloc[:,0:4]
y=irisdata.iloc[:,4]
a=y.unique()
a
```

```python
from sklearn import preprocessing 
y=pd.DataFrame(y)
le=preprocessing.LabelEncoder()
y=y.apply(le.fit_transform)
y.Target.unique()
y
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test ,y_train ,y_test =train_test_split(x,y, test_size=0.20)
y_train
```

```python
x_test
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train =scaler.transform(x_train)
x_test=scaler.transform(x_test)
x_train
```

```python
from sklearn.neural_network import MLPClassifier
mlp =MLPClassifier(hidden_layer_sizes=(10,10),max_iter=2000)
```

```python
mlp.fit(x_train , y_train.values.ravel())
```

```python
predictions =mlp.predict(x_test)
predictions
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, predictions))
print(classification_report(y_test,predictions))
```
